```python
happy = data['happyScore']
income = data['avg_income']
ineq = data['income_inequality']
print(happy)
```

```python
import matplotlib.pyplot as plt

plt.xlabel('income')
plt.ylabel('happy score')
plt.scatter(income, happy, s = ineq*10, alpha = 0.25)
```

```python
#using K-means clustering to Interpret the Data

from sklearn.cluster import KMeans
import numpy as np

income_happy = np.column_stack((income,happy)) #prints income and happy columns side by side 
print(income_happy)
```

```python
km_res = KMeans(n_clusters = 3).fit(income_happy) 
#you can change the number of clusters as per your choice

clusters = km_res.cluster_centers_
```

```python
plt.scatter(income, happy)
plt.scatter(clusters[:,0], clusters[:,1], s = 500, alpha = 0.5)
```
