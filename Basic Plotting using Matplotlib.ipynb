```python
import numpy as np
import matplotlib.pyplot as plot

xs = [10,100,25,67,74]
ys = [125,26,66,1,10]
xys = [[10,125],[100,26],[25,66],[67,1],[74,10]]

#plotting the mean

mean = np.mean(xys,0)

plot.scatter(xs,ys)
plot.scatter(mean[0],mean[1])
```

```python
#plotting the standard deviation

import matplotlib.patches as patches

SD = np.std(xys,0)
ellipse = patches.Ellipse(mean, SD[0]*2, SD[1]*2, alpha = 0.25)

fig, graph = plot.subplots()
graph.scatter(xs, ys)
graph.scatter(mean[0], mean[1])
graph.add_patch(ellipse)

plot.show()
```

```python
#calculting the distance to the mean

d0 = np.linalg.norm(xys[0]-mean)
d1 = np.linalg.norm(xys[1]-mean)

print(d0)
print(d1)
```

```python
min = np.min(xys,0)
max = np.max(xys,0)
print(min, max)
```

```python
normed = (xys - min) / (max - min)
print(normed)
```

```python
#plotting Normalised Data

import matplotlib.patches as patches

mean = np.mean(normed,0)
SD = np.std(normed,0)

ellipse = patches.Ellipse(mean, SD[0]*2, SD[1]*2, alpha = 0.25)

fig, graph = plot.subplots()

graph.scatter(normed[:0], normed[:1])
graph.scatter(mean[0], mean[1])
graph.add_patch(ellipse)

plot.show()
```
